```python
import math
import numpy as np
import pandas as pd
import matplotlib as mplt
import matplotlib.pyplot as plt
```

Electric Field of Q:

$$ \vec{E}_Q = \frac{\vec{F}}{q} = k\frac{Q}{r^2}\hat{r}

```python
# Plot Force of field felt by unit charge
k = 1/(4*math.pi*8.85e-12)
```

k, Coulumb's constant is equal to:

$$ k =  8.99×10^9 N⋅m^2 /C^2 = \frac{1}{4\pi\epsilon_0} $$

Where $\epsilon_0 $ is the "permittivity of free space" = $\ 8.85×10^{-12} $ F/m (farads per meter)


```python
r = np.linspace(1, 25, 100)
Q = 1e-9

E_Q= k*(Q/r**2)

plt.plot(r, E_Q)
plt.show()
```

```python
Q = 1e-6 # micro coulombs
points  = 500
side    = 100 # width of matrix
spacing = side/points

x1 = y1 = side/2 # charge particle in center of grid

xi = np.empty([points,points], float)

for i in range(points):
  y = spacing*i
  for j in range(points):
    x = spacing*j

    dist = math.sqrt((x - x1)**2 + (y - y1)**2)
    if dist == 0:
      continue

    xi[i,j] = k*(Q/dist**2)

df = pd.DataFrame(xi)
df.to_csv("./matrix.csv")

plt.imshow(xi, cmap='jet', origin="lower", norm=mplt.colors.LogNorm())
plt.colorbar()
plt.show()
```

As a reminder, our important constants are:

$$\ k = 8.99\times10^9N\cdot m^2 /C^2  = \frac{1}{4\pi\epsilon}$$
$\epsilon = 8.85×10^{-12} $ F/m (farads per meter)

Where a farad is $\ C^2/N \cdot m \times \frac{1}{m} $ makes $\ \epsilon = C^2/N \cdot m^2$
